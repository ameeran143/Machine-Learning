```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# image = plt.imread('2018Proj1_train/2.14.png')
image = plt.imread('nemo.png')

plt.imshow(image)

gray = rgb2gray(image)

gray_r = gray.reshape(gray.shape[0] * gray.shape[1])
for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 1
    else:
        gray_r[i] = 0

print("hello")
gray = gray_r.reshape(gray.shape[0], gray.shape[1])
plt.imshow(gray, cmap='gray')

plt.show()
```

```python
# using laplace convovle filter to see how vertcal and horizontal lines are detected

grid = cv2.imread('/Users/aashirmeeran/Desktop/Machine-Learning/Image_Segementation/index.png')
gray = rgb2gray(grid)
print(gray.shape)
plt.imshow(grid)
laplace_kernel = np.array([np.array([1, 1, 1]), np.array([1, -8, 1]), np.array([1, 1, 1])])
print(laplace_kernel)

out_l = ndimage.convolve(gray, laplace_kernel, mode='reflect')
plt.imshow(out_l, cmap='gray')


plt.show()
```
